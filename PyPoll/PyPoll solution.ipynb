```python
import os
import csv

# Open and read the CSV
csvpath = os.path.join('Resources', 'election_data.csv')
with open(csvpath, newline='') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    csv_header = next(csvreader)
    print(f"Header: {csv_header}")
    
    # Initialize variables
    votes = []
    candidates = []

    # Read each row of data after the header
    for row in csvreader:
        votes.append(row[0])
        candidates.append(row[2])

    # Total number of votes
    total_votes = len(votes)
    
    # Initialize candidate vote counts
    candidate_votes = {
        "Charles Casper Stockham": 0,
        "Diana DeGette": 0,
        "Raymon Anthony Doane": 0
    }

    # Count votes for each candidate
    for candidate in candidates:
        if candidate in candidate_votes:
            candidate_votes[candidate] += 1

    # Calculate percentages and find the winner
    winner = ""
    highest_votes = 0
    results = []
    
    for candidate, count in candidate_votes.items():
        percentage = round((count / total_votes) * 100, 3)
        results.append(f"{candidate}: {percentage}% ({count})")
        if count > highest_votes:
            highest_votes = count
            winner = candidate

    # Print Election Results
    print(f'''Election Results
-----------------------------------
Total Votes: {total_votes}
-----------------------------------''')
    
    for result in results:
        print(result)
    
    print(f'''-----------------------------------
Winner: {winner}
-----------------------------------''')

    # Output to a text file
    with open("output.txt", "w") as file:
        file.write(f'''Election Results
-----------------------------------
Total Votes: {total_votes}
-----------------------------------\n''')
        
        for result in results:
            file.write(f"{result}\n")
        
        file.write(f'''-----------------------------------
Winner: {winner}
-----------------------------------''')
```
